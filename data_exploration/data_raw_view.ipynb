# Wiley

```python
import glob

# Pega todos os arquivos txt (ajuste o caminho se necessário)
arquivos = glob.glob("C:/Users/Hychiro/Documents/Mestrado/Mestrado/Mestrado/Review/SLR-BCI-temporal-filters/raw_data/wiley_export_issue_*_Mar_18-2026.txt")

print(f"Arquivos encontrados: {len(arquivos)}")
print("Arquivos:")
for arquivo in arquivos:
    print(arquivo) 
```

```python
artigos = []
for arquivo in arquivos:
    with open(arquivo, "r", encoding="utf-8") as f:
        dados = {}
        for linha in f:
            linha = linha.strip()
            if not linha:
                continue
            if " - " in linha:
                chave, valor = linha.split(" - ", 1)
                # Se já existe a chave, acumula em lista (ex: vários autores)
                if chave in dados:
                    if isinstance(dados[chave], list):
                        dados[chave].append(valor)
                    else:
                        dados[chave] = [dados[chave], valor]
                else:
                    dados[chave] = valor
            # ER marca o fim do artigo
            if linha.startswith("ER"):
                # Normaliza listas em strings separadas por ";"
                for k, v in dados.items():
                    if isinstance(v, list):
                        dados[k] = "; ".join(v)
                artigos.append(dados)
                dados = {}

# Agora 'artigos' é uma lista de dicionários, cada um com os campos de um artigo
print(f"Foram lidos {len(artigos)} artigos.")
print(artigos[0])  # exemplo: mostra o primeiro artigo
```

```python
print("Campos disponíveis:")
if artigos:
    for campo in artigos[0].keys():
        print(campo)
```

```python
import pandas as pd

# Dropping unnecessary fields and creating a dataframe using only AU, TI, VL, PY
dfWiley = pd.DataFrame(artigos)

print("Campos antes da limpeza:")
print(dfWiley.columns)

# Remove espaços extras dos nomes das colunas
dfWiley.columns = dfWiley.columns.str.strip()

# Mantém apenas os campos desejados
dfWiley = dfWiley[["AU", "TI", "VL", "PY"]]

print("Campos depois da limpeza:")
print(dfWiley.columns)

# Renomeia colunas
dfWiley = dfWiley.rename(columns={
    "AU": "Autores",
    "TI": "Titulo",
    "VL": "Ano do Volume",
    "PY": "Ano de Publicacao"
})

print(dfWiley.head())

```

```python
print(dfWiley.info())
```

# IEEE

```python
import pandas as pd

# Caminho do arquivo
caminho = r"C:/Users/Hychiro/Documents/Mestrado/Mestrado/Mestrado/Review/SLR-BCI-temporal-filters/raw_data/ieee_export_Mar_18-2026.csv"

# Lê o CSV
dfIEEE = pd.read_csv(caminho, encoding="utf-8")

# Mostra informações básicas
print("Dimensões:", dfIEEE.shape)
```

```python
print(dfIEEE.head())
```

```python
# Mantém apenas as colunas desejadas
colunas_desejadas = ['Document Title', 'Authors', 'Publication Year', 'Issue Date']
dfIEEE_reduzido = dfIEEE.reindex(columns=colunas_desejadas)

# Renomeia para nomes mais claros
dfIEEE_reduzido = dfIEEE_reduzido.rename(columns={
    'Document Title': 'Titulo',
    'Authors': 'Autores',
    'Publication Year': 'Ano de Publicacão',
    'Issue Date': 'Ano do Volume'
})

print("DataFrame final:")
print(dfIEEE_reduzido.head())
```
